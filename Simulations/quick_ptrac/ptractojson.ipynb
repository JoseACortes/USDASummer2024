```python
import mcnptools as mctl
```

```python
ptracs = mctl.Ptrac('ptrac', mctl.Ptrac.ASC_PTRAC)
```

```python
histories = ptracs.ReadHistories(10000)
```

```python
len(histories)
```

```python
histories[0].GetNumEvents()
```

```python
ptrac_dict = []
particle_count = 1
for hist in histories:
    for i in range(hist.GetNumEvents()):
        event = hist.GetEvent(i)
        type = event.Type()
        x = event.Get(ptracs.X)
        y = event.Get(ptracs.Y)
        z = event.Get(ptracs.Z)
        time = event.Get(ptracs.TIME)
        particle = particle_count
        if type == ptracs.TER:
            particle_count += 1
        ptrac_dict.append({'particle': particle, 'type': type, 'x': x, 'y': y, 'z': z, 'time': time})

ptrac_dict
    
```

```python
len(ptrac_dict)
```

```python
# put into json
import json
with open('ptrac.json', 'w') as f:
    json.dump(ptrac_dict[:500], f)
    f.close()
```
