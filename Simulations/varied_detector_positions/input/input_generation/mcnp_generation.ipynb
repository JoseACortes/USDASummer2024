```python
def place_detector(x, y, z):
    detector_1_surface = {
        'surface_id': 2,
        'group': 2,
        'shape': 'right-box',
        'params': {
            'pos': (x, y+14.7, z),
            'size': (12.7, 12.7, 15.24),
        },
    }

    detector_2_surface = {
        'surface_id': 3,
        'group': 2,
        'shape': 'right-box',
        'params': {
            'pos': (x, y, z),
            'size': (12.7, 12.7, 15.24),
        },
    }
    detector_3_surface = {
        'surface_id': 4,
        'group': 2,
        'shape': 'right-box',
        'params': {
            'pos': (x, y-14.7, z),
            'size': (12.7, 12.7, 15.24),
        },
    }
    return [detector_1_surface, detector_2_surface, detector_3_surface]
```

```python
def place_shield(shield_id, group=4, x=0, y=0, z=0, h2_mult=1):
    Front_Pb_shield_surface = {
        'surface_id': shield_id,#5
        'group': group,
        'shape': 'right-box',
        'params': {
            'pos': (x+37.5, y+0, z-5),
            'size': (5, 60.1, 20),
        }
    }
    Front_Fe_shield_1_surface = {
        'surface_id': shield_id+1,#6
        'group': group,
        'shape': 'right-box',
        'params': {
            'pos': (x+32.5, y+0, z-5),
            'size': (5, 55, 20),
        }
    }
    Front_Ba_shield_surface = {
        'surface_id': shield_id+2,#7
        'group': group,
        'shape': 'right-box',
        'params': {
            'pos': (x+25, y+0, z-5),
            'size': (10, 55, 20),
        }
    }
    Front_Fe_shield_2_surface = {
        'surface_id': shield_id+3,#8
        'group': group,
        'shape': 'right-box',
        'params': {
            'pos': (x+17.5, y+0, z-5),
            'size': (5, 55, 20),
        }
    }
    Right_Fe_shield_surface = {
        'surface_id': shield_id+4,#9
        'group': group,
        'shape': 'right-box',
        'params': {
            'pos': (x-8, y+16.25, z-5),
            'size': (46, 2.5, 20),
        }
    }
    Right_Ba_shield_surface = {
        'surface_id': shield_id+5,#10
        'group': group,
        'shape': 'right-box',
        'params': {
            'pos': (x-8, y+22.5, z-5),
            'size': (46, 10, 20),
        }
    }
    Left_Fe_shield_surface = {
        'surface_id': shield_id+6,#11
        'group': group,
        'shape': 'right-box',
        'params': {
            'pos': (x-8, y-16.25, z-5),
            'size': (46, 2.5, 20),
        }
    }
    Left_Ba_shield_surface = {
        'surface_id': shield_id+7,#12
        'group': group,
        'shape': 'right-box',
        'params': {
            'pos': (x-8, y-22.5, z-5),
            'size': (46, 10, 20),
        }
    }
    return [Front_Pb_shield_surface, Front_Fe_shield_1_surface, Front_Ba_shield_surface, Front_Fe_shield_2_surface, Right_Fe_shield_surface, Right_Ba_shield_surface, Left_Fe_shield_surface, Left_Ba_shield_surface]
```

```python
def place_wheel(wheel_id, group, x=0, y=0, z=0, h2_mult=1):
    wheel_radius = 15.24
    tire_radius = 29
    tire_width = 25
    tread_thickness = 1.3
    tire_surface = {
        'surface_id': wheel_id,
        'group': group,
        'shape': 'cylinder',
        'params': {
            'pos': (x, y, z),
            'height': tire_width,
            'dir': (0,h2_mult,0),
            'r': tire_radius,
        }
    }
    wheel_surface = {
        'surface_id': wheel_id+1,
        'group': group,
        'shape': 'cylinder',
        'params': {
            'pos': (x, y, z),
            'height': tire_width,
            'dir': (0,h2_mult,0),
            'r': wheel_radius,
        }
    }
    inside_tire_surface = {
        'surface_id': wheel_id+2,
        'group': group,
        'shape': 'cylinder',
        'params': {
            'pos': (x, y+(tread_thickness*h2_mult), z),
            'height': (tire_width-tread_thickness*2),
            'dir': (0,h2_mult,0),
            'r': tire_radius-tread_thickness,
        }
    }
    return [tire_surface, wheel_surface, inside_tire_surface]
```

# MCNP Generation

```python
def sphere_to_mcnp(sphere):
    # MCNP sphere is defined by the center and the radius
    id = sphere['surface_id']
    group = sphere['group']
    pos = sphere['params']['pos']
    r = sphere['params']['r']

    return f"{id} sph {pos[0]} {pos[1]} {pos[2]} {r}"

def box_to_mcnp(box):
    # current box is defined in the vpython method, by the center, the axis, and the size
    # mcnp box is defined by the a corner and 3 orthogonal vectors that define the box
    # the corner is the center - 0.5*size
    # the vectors are the axis scaled by the size

    id = box['surface_id']
    group = box['group']
    pos = box['params']['pos']
    size = box['params']['size']

    corner = [pos[i] - 0.5*size[i] for i in range(3)]
    vectors = [[0,0,0] for i in range(3)]
    for i in range(3):
        vectors[i][i] = size[i]

    
    return f"{id} box {corner[0]} {corner[1]} {corner[2]}   {vectors[0][0]} {vectors[0][1]} {vectors[0][2]}   {vectors[1][0]} {vectors[1][1]} {vectors[1][2]}   {vectors[2][0]} {vectors[2][1]} {vectors[2][2]}"

def cylinder_to_mcnp(cylinder):

    id = cylinder['surface_id']
    group = cylinder['group']
    pos = cylinder['params']['pos']
    dir = cylinder['params']['dir']
    r = cylinder['params']['r']
    height = cylinder['params']['height']

    return f"{id} rcc {pos[0]} {pos[1]} {pos[2]}    {height*dir[0]} {height*dir[1]} {height*dir[2]} {r}"

def generate_surfaces(surfaces):
    mcnp = []
    for surface in surfaces:
        if surface['shape'] == 'sphere':
            mcnp.append(sphere_to_mcnp(surface))
        elif surface['shape'] == 'right-box':
            mcnp.append(box_to_mcnp(surface))
        elif surface['shape'] == 'cylinder':
            mcnp.append(cylinder_to_mcnp(surface))
    return mcnp
```

```python
def generate_cells(cells):
    mcnp = []
    for cell in cells:
        if cell['material']==0:
            mcnp.append(f"{cell['cell_id']} {cell['material']} {cell['surfaces']} imp:n,p,e 0")
        else:
            mcnp.append(f"{cell['cell_id']} {cell['material']} {cell['density']} {cell['surfaces']} imp:n,p,e 1")
    return mcnp
```

```python
def generate_material(material):
    mat = []
    for key, value in material["composition"].items():
        mat.append(f"\t{key}\t{value}")
    mat[0] = f"m{material['id']}"+mat[0]
    if 'hydration' in material:
        hydration = str(material['hydration'])[1:]
        mat.append(f"MT{material['id']} lwtr{hydration}")
    return mat
def generate_materials(materials):
    mcnp = []
    for material in materials:
        mcnp += generate_material(material)
    return mcnp
```

```python
def source_to_mcnp(source):
    pos = source['pos']
    pos = ' '.join([str(i) for i in pos])
    erg = source['erg']
    wgt = source['wgt']
    vec = source['vec']
    vec = ' '.join([str(i) for i in vec])
    return [f'SDEF POS={pos} ERG={erg} WGT={wgt} VEC={vec}']
```

```python
def tally_to_mcnp(tally):
    id = tally['tally_id']
    tally_string = f"F{id}:{tally['particle']} {tally['cells']} "
    mcnp = [tally_string]
    if 'energy_bins' in tally: 
        tally_string = f"e{id} {tally['energy_bins']} "
        mcnp.append(tally_string)
    if 'time_bins' in tally:
        tally_string = f"T{id} {tally['time_bins']} "
        mcnp.append(tally_string)
    # if any of the following are in the tally then do something
    if ('phl' in tally) or ('geb' in tally):
        tally_string = f"FT{id} "
        if 'phl' in tally:
            phl_string = "phl "
            for phl in tally['phl']:
                phl_string += f"{len(phl['tallies'])} "
                for _ in range(len(phl['tallies'])):
                    phl_string += f"{phl['tallies'][_]} {phl['fbins'][_]} "
                phl_string += "0 "
            tally_string += phl_string
        if 'geb' in tally:
            tally_string += f"geb {tally['geb']['a']} {tally['geb']['b']} {tally['geb']['c']} "
        if 'tmc' in tally:
            tally_string += f"tmc {tally['tmc']['a']} {tally['tmc']['b']} "
        mcnp.append(tally_string)
    return mcnp

def tallies_to_mcnp(tallies):
    mcnp = []
    for tally in tallies:
        mcnp += tally_to_mcnp(tally)
    return mcnp
```
